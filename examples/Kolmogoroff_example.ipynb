```julia
using PyPlot, Seismic

dt = 0.002
w = Ricker(dt=dt)
wmin = SeisKolmogoroff(w)

nw = length(w)
t = dt*collect(0:1:nw-1)

figure(figsize=(6, 6))
subplots_adjust(hspace = 0.4)
subplot(211)
plot(t, w)
axis("tight")

title("Ricker wavelet")

subplot(212)
plot(t,wmin)
xlabel("Time (s)")
title("Minimum phase equivalent")
```

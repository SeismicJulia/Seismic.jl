```julia
using Seismic, SeismicPlotting
```

```julia
download("http://seismic.physics.ualberta.ca/data/sigmoid.rsf","sigmoid.rsf");
download("http://seismic.physics.ualberta.ca/data/sigmoid.rsf@","sigmoid.rsf@");
SegyToSeis("sigmoid.rsf","sigmoid",["format"=>"rsf"]);
```

```julia
d,h = SeisRead("sigmoid");
plotpar = ["style"=>"wiggles",
           "vmin"=>-1e-3,"vmax"=>1e-3,
           "aspect"=>"auto",
           "xlabel"=>"X","xunits"=>"","ox"=>0,"dx"=>1,
           "ylabel"=>"Time","yunits"=>"(seconds)","oy"=>0,"dy"=>h[1].d1,
           "cmap"=>"gray",
           "title"=>"sigmoid.rsf"];
SeisPlot(d[:,:],plotpar);
```

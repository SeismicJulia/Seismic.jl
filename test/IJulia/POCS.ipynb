```julia
using PyPlot,Seismic
```

```julia
param = {"nt"=>500,"dt"=>0.004,"nx1"=>500,
"tau1"=>[0.4 1.0],"tau2"=>[0. 0.],"tau3"=>[0. 0.],"tau4"=>[0. 0.],
"v1"=>[3500. -4000],"v2"=>[99999. 99999.],"v3"=>[99999. 99999.],"v4"=>[99999. 99999.],
"amp"=>[1. -0.5], "f0"=>[20. 20.]};
d = SeisLinearEvents(param); 
d = d[:,:];
ddec = copy(d)
for ix = 1:size(d,2)
    if (rand() < 0.5); ddec[:,ix] = 0.0; end;
end
```

```julia
param = {"Niter"=>100,"fmax"=>60,"padt"=>2,"padx"=>2,"dt"=>0.004}
dpocs = Seismic.pocs(ddec,param);
```

```julia
figure(figsize=(10, 5))
subplot(121)
SeisPlot(ddec,{"cmap"=>"seismic","fignum"=>1})
subplot(122)
SeisPlotFKSpectrum(ddec,{"dy"=>0.004,"fignum"=>1,"hbox"=>5})
```

```julia
figure(figsize=(10, 5))
subplot(121)
SeisPlot(dpocs,{"cmap"=>"seismic","fignum"=>1})
subplot(122)
SeisPlotFKSpectrum(dpocs,{"dy"=>0.004,"fignum"=>1})
```
